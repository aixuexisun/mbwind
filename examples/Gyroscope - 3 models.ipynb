Test modelling of a gyroscope and precession/nutation.

System consists of 3 hinges at origin about 3 axes to allow the gyroscope to
spin, tilt in elevation and rotate in azimuth.

The rigid spinning body is modelled in 3 ways:

 1. RigidBody element, mass and inertia specified directly.
 2. ModalElement element with no mode shapes, cross-section inertia and
    density distribution given.
 3. UniformBeam element with rigid strains, density and extra cross-section
    inertia given.

Make some specific gyroscope models, which use different types of element to represent the rotor:

```python
class RigidBodyGyroscope(Gyroscope):
    def build_body(self):
        Jx = self.radius**2 / 2
        Jyz = (3*self.radius**2 + self.length**2) / 12
        Jyz_0 = Jyz + (self.length/2)**2 # parallel axis theorem
        inertia = mass * np.diag([Jx, Jyz_0, Jyz_0])
        return RigidBody('body', mass, inertia, [length/2, 0, 0])

    
class BeamGyroscope(Gyroscope):
    def build_body(self):
        Jx = self.mass * self.radius**2 / 2
        return UniformBeam('body', self.length, self.mass/self.length,
                           1, 1, 1, Jx=Jx/2) # /2 because added to both ends

    def prescribe_body(self):
        self.system.prescribe(self.body, vel=0.0) # rigid beam

class ModalGyroscope(Gyroscope):
    def build_body(self):    
        x = np.linspace(0, self.length)
        modes = ModalRepresentation(
            x             =x,
            shapes        =np.zeros((len(x),3,0)),
            rotations     =np.zeros((len(x),3,0)),
            density       =np.ones_like(x) * mass/length,
            mass_axis     =np.zeros((len(x),2)),
            section_inertia=np.ones_like(x) * radius**2 / 4,
            freqs         =np.zeros((0,)),
        )
        return ModalElement('body', modes)
```

Create 3 different models

```python
length = 3.0
radius = 1.0
mass = 100.0
spin = 30.0

bg = BeamGyroscope(length, radius, mass, spin)
mg = ModalGyroscope(length, radius, mass, spin)
rg = RigidBodyGyroscope(length, radius, mass, spin)
```

```python
def plot_position(gyro, title=None):
    # Theory
    spin_inertia = gyro.mass * gyro.radius**2 / 2
    cross_inertia = gyro.mass * (3*gyro.radius**2 + gyro.length**2) / 12
    spin_velocity = gyro.spin
    torque = gyro.mass * gravity * gyro.length/2
    precession = torque / (spin_inertia * spin_velocity)
    
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax2 = ax1.twinx()
    
    ax1.plot(gyro.t, np.degrees(gyro.y[0] % (2*np.pi)), 'b', label='Azimuth')
    ax1.plot(gyro.t, np.degrees((gyro.t*precession) % (2*np.pi)), 'b:')
    #ax1.plot(gyro.t, np.degrees(gyro.y[2] % (2*np.pi)), 'g', label='Spin')
    ax1.set_ylim((-10,360))            
    ax1.legend()
    ax1.set_ylabel('Azimuth / deg')
    
    ax2.plot(gyro.t, -np.degrees(gyro.y[1] % (2*np.pi)), 'r', label='Elevation')
    ax2.set_ylim((-90,10))
    ax2.set_ylabel('Elevation / deg')
    
    ax1.set_xlabel('Time / s')
    if title is not None:
        ax1.set_title(title)
        
def plot_velocities(gyro, title=None):
    # Theory
    spin_inertia = gyro.mass * gyro.radius**2 / 2
    cross_inertia = gyro.mass * (3*gyro.radius**2 + gyro.length**2) / 12
    spin_velocity = gyro.spin
    torque = gyro.mass * gravity * gyro.length/2
    precession = torque / (spin_inertia * spin_velocity)
    nutation = cross_inertia * spin_velocity / spin_inertia
    nutamp = 1
    
    # Skip body DOFs if necessary
    if gyro.body.__class__ is UniformBeam:
        angvels = gyro.y[4:6]
    else:
        angvels = gyro.y[3:5]
    
    fig, ax = plt.subplots()
    
    ax.plot(gyro.t, angvels[0], 'b', label='d/dt azimuth')
    ax.plot(gyro.t, np.ones_like(gyro.t)*precession, 'b:')
    
    ax.plot(gyro.t, angvels[1], 'r', label='d/dt elevation')
    #ax.plot(gyro.t, nutamp*np.sin(nutation*gyro.t), 'r--')
    
    ax.legend()
    ax.set_ylabel('rad/s')
    ax.set_xlabel('Time / s')
    if title is not None: ax.set_title(title)
```
