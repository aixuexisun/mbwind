Simulate an increasing wind load on a flexible beam, and compare to results calculated by the wind turbine simulation software Bladed.

The coordinate systems are related as follows.

    Bladed x-deflection = -z deflection here. 
    Bladed y-deflection = +y deflection here.
    Bladed OOP force = -z loading here.
    Bladed IP force  = +y loading here.

```python
%matplotlib inline
import numpy as np
from scipy import interpolate
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from mbwind import System, ModalElementFromFE, Integrator, CustomOutput
from sample_blade import sample_blade_fe

blade, fe = sample_blade_fe()

# Blade loading - force in -z direction
wind_table = np.array([
    [0, 1,   2,  10],  # time
    [0, 0,   0,   0],  # x force
    [0, 0,   0,   0],  # y force
    [0, 0, -20, -20],  # z force
])
wind_interp = interpolate.interp1d(wind_table[0], wind_table[1:])

def loading_func(el, t):
    # Find relative wind speed
    wind_vel = wind_interp(t)
    element_vel = el.elastic_velocities()
    rel_vel = wind_vel[None, :] - element_vel
    
    # Calculate loading for each station.
    # In Bladed, last station has zero loading
    Cd = 0.5     # Drag coefficient
    rho = 1.225  # Air density
    force = np.zeros((len(element_vel), 3))
    for i in range(len(element_vel) - 1):
        rel_dir = np.arctan2(rel_vel[i, 2], rel_vel[i, 1])
        rel_speed = np.sqrt((rel_vel[i, 1] + rel_vel[i, 2])**2)        
        diameter = blade['chord'][i]
        dragforce = 0.5 * rho * Cd * diameter * rel_speed**2
        force[i, :] = dragforce * array([0, np.cos(rel_dir), np.sin(rel_dir)])
    return force
```

Setup the mbwind model:

```python
# Modal element
el = ModalElementFromFE('el', fe, num_modes=4, damping=0.005)
el.loading = loading_func
system = System()
system.add_leaf(el)
system.setup()
```

Intergrate results:

```python
integ = Integrator(system)
integ.add_output(el.output_deflections(stations=[8,16]))
integ.add_output(CustomOutput(lambda s: el.loading(el, s.time)[8,:], 'loading'))
t, y = integ.integrate(10, 0.05)
integ.labels
```

Save results for later use:

```python
np.savez('blade_nrel_parked_mbwind_results.npz', t=t, 
         strains=y[0], 
         deflections=y[1], 
         loading=y[2], 
         labels=integ.labels)
```

Load Bladed results for comparison:

```python
bladed_results = np.load('Bladed_models/blade_nrel_parked_bladed_results.npy')
```

```python
def plot(t, y):
    fig = plt.figure(figsize=(9, 8))
    gs = gridspec.GridSpec(4, 2)

    ax = fig.add_subplot(gs[0, 0])
    ax.plot(wind_table[0], -wind_table[3].T)
    ax.set_ylabel('Wind speed (m/s)')
    ax.set_ylim(0, 25)
    
    ax = fig.add_subplot(gs[1, 0])
    ax.plot(t, -y[2][:, 2], 'b', label='Mine')
    ax.plot(t, bladed_results['dfout1'], 'k--', label='Bladed')
    ax.set_ylabel('OOP force (N/m)')
    ax.set_ylim(0, 600)
    
    ax = fig.add_subplot(gs[1, 1])
    ax.plot(t, y[2][:, 1], 'b', label='Mine')
    ax.plot(t, bladed_results['dfin1'], 'k--', label='Bladed')
    ax.set_ylabel('IP force (N/m)')
    
    ax = fig.add_subplot(gs[2:4, 0])
    ax.plot(t, -100*y[1][:, 0, 2], 'b', label='Mine (midspan)')
    ax.plot(t, -100*y[1][:, 1, 2], 'g', label='Mine (tip)')
    ax.plot(t, 100*bladed_results['x1'], 'k--', label='Bladed')
    ax.plot(t, 100*bladed_results['x2'], 'k--')
    ax.set_ylabel('OOP defl (cm)')
    ax.set_xlabel('Time / s')
    ax.legend(frameon=False, fontsize='small')
    
    ax = fig.add_subplot(gs[2:4, 1])
    ax.plot(t, 100*y[1][:, 0, 1], 'b', label='Mine')
    ax.plot(t, 100*y[1][:, 1, 1], 'g')
    ax.plot(t, 100*bladed_results['y1'], 'k--', label='Bladed')
    ax.plot(t, 100*bladed_results['y2'], 'k--')
    ax.set_ylabel('IP defl (cm)')
    ax.set_xlabel('Time / s')
plot(t, y)
```
