Test joint reaction forces: swinging pendulum with lumped mass

```python
from __future__ import division

import numpy as np
import matplotlib.pyplot as plt

import mbwind
from mbwind import (System, RigidBody, Hinge, RigidConnection, Integrator, LoadOutput, CustomOutput)

# Parameters
mass = 11.234
length = 2.54
gravity = 9.81

# Build model
hinge = Hinge('hinge', [0, 1, 0])
link = RigidConnection('link', [length, 0, 0])
body = RigidBody('body', mass)

system = System(gravity=gravity)
system.add_leaf(hinge)
hinge.add_leaf(link)
link.add_leaf(body)
system.setup()
```

To check the results are correct, define some custom outputs which calculate the expected reaction forces based on the simulated velocities of the mass. Do this for two points:

 1. At the pendulum bob, in local coordinates (aligned with the link)
 2. At the ground attachment, in global coordinates

```python
# Custom outputs to calculate correct answer
def force_body_prox_local(s):
    theta = s.q[hinge.istrain][0]
    thetadot = s.qd[hinge.istrain][0]
    thetadotdot = s.qdd[hinge.istrain][0]
    Fx = mass * (-gravity*np.sin(theta) - length*thetadot**2)
    Fz = mass * ( gravity*np.cos(theta) - length*thetadotdot)
    return [Fx, 0, Fz, 0, 0, 0]

def force_hinge_prox(s):
    theta = s.q[hinge.istrain][0]
    thetadot = s.qd[hinge.istrain][0]
    thetadotdot = s.qdd[hinge.istrain][0]
    A = np.array([[ np.cos(theta), np.sin(theta)],
                  [-np.sin(theta), np.cos(theta)]])
    Fxz = -mass * length * np.dot(A, [thetadot**2, thetadotdot])
    return [Fxz[0], 0, Fxz[1] + gravity*mass, 0, 0, 0]
```

Set up the integrator and run the simulation:

```python
# Solver
integ = Integrator(system, ('pos','vel','acc'))
integ.add_output(LoadOutput(hinge.iprox))
integ.add_output(LoadOutput(link.iprox))
integ.add_output(LoadOutput(body.iprox))
integ.add_output(LoadOutput(body.iprox, local=True))
integ.add_output(CustomOutput(force_hinge_prox, "correct ground"))
integ.add_output(CustomOutput(force_body_prox_local, "correct link distal local"))

# Run a simluation
t, y = integ.integrate(3.8, 0.03)
integ.labels
```

```python
fig = plt.figure(figsize=(8, 9))

ax = fig.add_subplot(321)
ax.plot(t, y[0][:, 0] * 180 / pi)
ax.set_title('Rotation [deg]')

# reaction forces vs predictions - base
ax = fig.add_subplot(323)
ax.set_color_cycle(['r','g','b'])
ax.plot(t, y[3][:, :3], t, y[7][:, :3], 'k--')
ax.set_title('Base - forces')

# reaction forces vs predictions - base moments
ax = fig.add_subplot(324)
ax.set_color_cycle(['r','g','b'])
ax.plot(t, y[3][:, 3:], t, y[7][:, 3:], 'k--')
ax.set_title('- moments')

# reaction forces vs predictions - local body prox
ax = fig.add_subplot(325)
ax.set_color_cycle(['r','g','b'])
ax.plot(t, y[6][:, :3], t, y[8][:, :3], 'k--')
ax.set_title('Body local - forces')

# reaction forces vs predictions - base of link moments
ax = fig.add_subplot(326)
ax.set_color_cycle(['r','g','b'])
ax.plot(t, y[6][:, 3:], t, y[8][:, 3:], 'k--')
ax.set_title('- moments')

ax.legend(('x','y','z'));
```

Note that the force at the mass is purely axial (as expected).
