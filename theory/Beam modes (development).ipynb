# Step 1 - FE model of beam

Make a simple FE model of a beam -- define the shape functions:

```python
import sympy
%load_ext sympyprinting
%pylab inline
```

```python
x, x1, x2, syml, EI, rho = sympy.symbols('x, x_1, x_2, l, EI, rho')
l = x2 - x1
N1 = -(x - x2)**2 * (-l + 2*(x1 - x)) / l**3
N2 =  (x - x1)**2 * ( l + 2*(x2 - x)) / l**3
M1 = (x - x1)**1 * (x - x2)**2 / l**2
M2 = (x - x1)**2 * (x - x2)**1 / l**2

d2dx_N1 = N1.diff(x, 2).expand().simplify()
d2dx_N2 = N2.diff(x, 2).expand().simplify()
d2dx_M1 = M1.diff(x, 2).expand().simplify()
d2dx_M2 = M2.diff(x, 2).expand().simplify()

xi = sympy.symbols('xi')
display(d2dx_N1.subs(x, xi*l + x1).subs(4*x2 - 4*x1, 4*l).expand().simplify())
display(d2dx_N2.subs(x, xi*l + x1).subs(2*x2 - 2*x1, 2*l).expand().simplify())
display(d2dx_M1.subs(x, xi*l + x1).subs(2*x2 - 2*x1, 2*l).expand().simplify())
display(d2dx_M2.subs(x, xi*l + x1).subs(1*x2 - 1*x1, 1*l).expand().simplify())
```

```python
def calc_B(x1_value, x2_value):
    # From 3D7 notes page 30
    l_value = x2_value - x1_value
    B = array([dd.subs({x1: x1_value, x2: x2_value, l: l_value})
               for dd in (d2dx_N1, d2dx_M1, d2dx_N2, d2dx_M2)])
    return B

def sym_ke():
    B = array([[d2dx_N1, d2dx_M1, d2dx_N2, d2dx_M2]])
    integrand = dot(B.T, B)
    ke = zeros_like(integrand)
    for i,xxx in enumerate(integrand.flat):
        ke.flat[i] = sympy.integrate(EI * xxx, (x, x1, x2)).expand().simplify()
    return ke

def sym_me():
    N = array([[N1, M1, N2, M2]])
    integrand = dot(N.T, N)
    me = zeros_like(integrand)
    for i,xxx in enumerate(integrand.flat):
        me.flat[i] = sympy.integrate(rho * xxx, (x, x1, x2)).expand().simplify()
    return me

def be_point_load(xi, fi):
    N = array([dd.subs({x: xi}).simplify() for dd in (N1, M1, N2, M2)])
    be = N.T * fi
    return be
```

```python
B = calc_B(0, 1)
sympy.plot(B[0], B[1], B[2], B[3], (x, 0, 1))
```

```python
ke = sym_ke()
sympy.Matrix(array([ki.factor().subs({(x2 - x1): syml})/EI*syml**3 for ki in ke.flat]).reshape((4,4)))
```

```python
me = sym_me()
sympy.Matrix(array([mi.factor().subs({(x2 - x1): syml})/rho for mi in me.flat]).reshape((4,4)))
```

### Assemble model

Define the number of elements and the numerical values

```python
N_elements = 30
EI_value = 1e3
rho_value = 7800 * pi * 0.1**2  # mass per unit length
total_length = 60.0
```

```python
dx = total_length / N_elements
N_nodes = N_elements + 1
K = zeros((N_nodes*2, N_nodes*2))
M = zeros((N_nodes*2, N_nodes*2))
b = zeros(N_nodes*2)
for iel in range(N_elements):
    for i in range(4):
        for j in range(4):
            I = 2*iel + i
            J = 2*iel + j
            K[I,J] += ke[i,j].subs({x1: iel*dx, x2: (iel+1)*dx, EI: EI_value})
            M[I,J] += me[i,j].subs({x1: iel*dx, x2: (iel+1)*dx, rho: rho_value})
matshow(K); colorbar()
matshow(M); colorbar();
```

```python
from scipy.linalg import eig
w, v = eig(K, M)
order = argsort(w)
w = sqrt(w[order].real)
v = v[:,order]
bar(range(len(w)), w)
```

# Step 2 - substructuring

To use the Craig-Bampton method, two things are needed:

  1) the static boundary modes, $\boldsymbol{\Xi} = 
    \begin{bmatrix} \boldsymbol{I} \\ -\boldsymbol{K}_{II}^{-1}\boldsymbol{K}_{IB} \end{bmatrix}$
    
  2) a reduced set of $r$ normal modes, $\boldsymbol{\Phi}$
    
First, we'll split the nodes into boundary (B) nodes, at the ends of the beam, and interior (I) nodes:

```python
boundary_dofs = [0, 1, K.shape[0]-2, K.shape[0]-1]
idx_B = zeros(K.shape[0], dtype=bool)
idx_B[boundary_dofs] = True
K_BB = K[ idx_B, :][:,  idx_B]
K_BI = K[ idx_B, :][:, ~idx_B]
K_II = K[~idx_B, :][:, ~idx_B]
K_IB = K[~idx_B, :][:,  idx_B]
```

Calculate the boundary modes:

```python
from numpy import linalg
Xi = zeros((K.shape[0], len(boundary_dofs)))
Xi[~idx_B, :] = -dot(linalg.inv(K_II), K_IB)
Xi[ idx_B, :] = eye(len(boundary_dofs))
Xi.shape
```

```python
plot(Xi[0::2])
title('Internal deflections due to unit boundary deflections')
legend(boundary_dofs);
```

The normal modes have already been calculated above (the clamped-clamped eigenvectors). Let's take the first 3 modes, say:

```python
r = 3  # number of normal modes
Phi = zeros((K.shape[0], r))
Phi[~idx_B, :] = v[2:-2, :r]
Phi[0] *= -1
Phi.shape
```

```python
plot(Phi[0::2])
title('Internal deflections due to normal modes')
legend(range(r));
```

Combined, these give the Craig-Bampton reduction matrix:

```python
CB_reduction = c_[ Xi, Phi ]
```

From this we get the statically-reduced stiffness and mass matrices:

```python
Kred = dot(CB_reduction.T, dot(K, CB_reduction))
Mred = dot(CB_reduction.T, dot(M, CB_reduction))
matshow(Kred)
matshow(Mred);
```

Check that the internal parts of the reduced matrices recover the natural frequencies:

```python
allclose(sqrt(diag(Kred[4:, 4:]) / diag(Mred[4:, 4:])), w[:3])
```

What do the boundary DOF parts of the matrices look like?

```python
print "Translation:"
print Mred[0:4:2, 0:4:2]
print "Rotation:"
print Mred[1:4:2, 1:4:2]

print "Translation:"
print Kred[0:4:2, 0:4:2]
print "Rotation:"
print Kred[1:4:2, 1:4:2]
```

The stiffness part makes sense -- the force at either end depends on the difference between the transverse deflection, and similarly with the rotation but there is a factor of 2 as expected from the structures databook.

There is no stiffness coupling between the boundary DOFs and the internal DOFs:

```python
allclose(Kred[:4, 4:], 0) and allclose(Kred[4:, :4], 0)
```

But there is mass coupling -- for example the translational parts:

```python
print Mred[0:4:2, 4:]
```

The two rows correspond to the inertial forces at each of the boundary nodes due to motion in the internal (normal) modes. The first mode is symmetric, so the force is the same at both ends; the second mode is antisymmetric, so the forces are opposite; and so on.

## Conclusions

This shows that:
    
  - I can correctly calculate the normal and constraint modes of a uniform beam, at least they match Bladed.
  - Bladed's "attachment modes" conflict with the definition in Geradin & Cardona, in that they are the shapes due to unit deflections rather than unit forces (G&C call these "constraint modes")
    
In mbwind, the boundary nodes are at the proximal and distal nodes. Since the element needs a reference coordinate system, I will choose this to be attached to the proximal node. The matrices above are all in local coordinates; therefore, before adding the mass matrix entries to the system matrix, I need to transform the coordinates into global coordinates.

Let the global position of a point be $\boldsymbol{r} = \boldsymbol{r}_0 + \boldsymbol{R}_0 (\boldsymbol{X} + \boldsymbol{u})$, where $\boldsymbol{r}_0$ and $\boldsymbol{R}_0$ are the position and orientation of the reference frame. The global orientation of the point is $\boldsymbol{\Psi} = \boldsymbol{\Psi}_0 \circ \boldsymbol{\psi}$, where $\circ$ indicates combinations of rotations (its arguments are rotation parameters, not rotation matrices).

Turning this around, the relative (elastic) deflection and rotation in the body is
$$
\begin{bmatrix} \boldsymbol{u} \\ \boldsymbol{\psi} \end{bmatrix} = 
\begin{bmatrix} \boldsymbol{R}_0^T (\boldsymbol{r} - \boldsymbol{r}_0) - \boldsymbol{X} \\
-\boldsymbol{\Psi}_0 \circ \boldsymbol{\Psi} \end{bmatrix}
$$

Let these deflections be given by some shape functions, in this case the Craig-Bampton ones:
$$
\begin{bmatrix} \boldsymbol{u} \\ \boldsymbol{\psi} \end{bmatrix} = 
\boldsymbol{\bar{\Phi}} \boldsymbol{y} =
\boldsymbol{\Xi}\boldsymbol{y}_B + \boldsymbol{\Phi}\boldsymbol{\eta} =
\boldsymbol{\Xi}\begin{bmatrix} \boldsymbol{u}_B \\ \boldsymbol{\psi}_B \end{bmatrix} + \boldsymbol{\Phi}\boldsymbol{\eta}
$$

The generalised coordinates "$\boldsymbol{q}$" of the superelement are the set of positions and rotations, in global coordinates, of all the boundary nodes $B$, together with the normal modal amplitudes:
$$
\boldsymbol{q} = \begin{bmatrix} \boldsymbol{r}_B \\ \boldsymbol{\Psi}_B \\ \boldsymbol{\eta} \end{bmatrix}
$$

So the problem is to relate the generalised coordinates $\boldsymbol{r}_B$ and $\boldsymbol{\Psi}_B$ to the static modal amplitudes $\boldsymbol{y}_B$.

# Step 4 - Coordinate transformations

The reduced matrices I have derived above are defined in terms of the local coordinates $\boldsymbol{y}$:
$$\boldsymbol{y} = \begin{bmatrix} \boldsymbol{u}_B \\ \boldsymbol{\psi}_B \\ \boldsymbol{\eta} \end{bmatrix}$$
(although not exactly in that order -- the deflections and rotations are interleaved). I'm assuming small rotations within the body frame, so I don't have to be so precise about how the rotations \boldsymbol{\psi}_B are defined -- they are just small rotations about each axis.

The deflection velocity and angular velocity at each boundary node are given by:
$$
\begin{bmatrix} \dot{\boldsymbol{u}}_B \\ \boldsymbol{\Omega}_B \end{bmatrix} = 
\begin{bmatrix} \boldsymbol{R}_0^T (\dot{\boldsymbol{r}}_B - \dot{\boldsymbol{r}}_0) +
  \boldsymbol{R}_0^T \widetilde{(\boldsymbol{r}_B - \boldsymbol{r}_0)}\boldsymbol{\omega}_0 \\
\boldsymbol{R}_0^T (\boldsymbol{\omega}_B - \boldsymbol{\omega}_0) \end{bmatrix}
$$

So
$$
\begin{bmatrix} \dot{\boldsymbol{u}}_p \\ \boldsymbol{\Omega}_p \\ \dot{\boldsymbol{u}}_d \\ \boldsymbol{\Omega}_d \end{bmatrix} = 
\begin{bmatrix}
\boldsymbol{0} & \boldsymbol{0} & \boldsymbol{0} & \boldsymbol{0} \\
\boldsymbol{0} & \boldsymbol{0} & \boldsymbol{0} & \boldsymbol{0} \\
-\boldsymbol{R}_p^T & \boldsymbol{R}_p^T \widetilde{(\boldsymbol{r}_d - \boldsymbol{r}_p)} & \boldsymbol{R}_p^T & \boldsymbol{0} \\
\boldsymbol{0} & -\boldsymbol{R}_p^T & \boldsymbol{0} & \boldsymbol{R}_p^T \\
\end{bmatrix}
\begin{bmatrix} \dot{\boldsymbol{r}}_p \\ \boldsymbol{\omega}_p \\ \dot{\boldsymbol{r}}_d \\ \boldsymbol{\omega}_d \end{bmatrix}
$$

Let's check this in a few cases.

```python
from mbwind.utils import skewmat, rotmat_z
def Ymat(Rp, rp, rd):
    rp, rd = asarray(rp), asarray(rd)
    Y = zeros((12, 12))
    Y[6:9, 0:3] = -Rp.T
    Y[6:9, 3:6] = dot(Rp.T, skewmat(rd - rp))
    Y[6:9, 6:9] = Rp.T
    Y[9:12, 3:6] = -Rp.T
    Y[9:12, 9:12] = Rp.T
    return Y
```

To transform the reduced mass and stiffness matrices into global coordinates, need to pad them out a bit to be three-dimensional -- make the modes be in the y-direction

Be careful because the attachment modes above are (y1, theta_z1, y2, theta_z2), but the order in the 3d matrix is (x1, y1, x1, theta_x1, theta_y1, theta_z1, x2, y2, z2, theta_x2, theta_y2, theta_z2).

```python
N3 = 6*2 + 6 # 12 rigid DOFs, 3 modes
def assemble(B):
    A = zeros((N3,N3))
    A[1:12:6, 1:12:6] = B[0:4:2, 0:4:2] # translations
    A[2:12:6, 2:12:6] = B[0:4:2, 0:4:2]
    A[5:12:6, 5:12:6] = B[1:4:2, 1:4:2] # rotations
    A[4:12:6, 4:12:6] = B[1:4:2, 1:4:2]
    
    # cross-terms
    A[1:12:6, 12::2] = B[:4:2, 4:] # translations
    A[2:12:6, 13::2] = B[:4:2, 4:]
    A[12::2, 1:12:6] = B[4:, :4:2]
    A[13::2, 2:12:6] = B[4:, :4:2]
    A[5:12:6, 12::2] = B[1:4:2, 4:] # rotations
    A[4:12:6, 13::2] = B[1:4:2, 4:]
    A[12::2, 5:12:6] = B[4:, 1:4:2]
    A[13::2, 4:12:6] = B[4:, 1:4:2]
    
    # modes x2
    A[12::2, 12::2] = B[4:, 4:]
    A[13::2, 13::2] = B[4:, 4:]

    return A

Kred3 = assemble(Kred)
Mred3 = assemble(Mred)
```

The deflections and rotations at the boundary nodes are given by:
$$
\begin{bmatrix} \boldsymbol{u}_B \\ \boldsymbol{\psi}_B \end{bmatrix} = 
\begin{bmatrix} \boldsymbol{R}_0^T (\boldsymbol{r}_B - \boldsymbol{r}_0) - \boldsymbol{X}_B \\
-\boldsymbol{\Psi}_0 \circ \boldsymbol{\Psi}_B \end{bmatrix}
$$

Let's now get specific to my element: $\boldsymbol{r}_B = \begin{bmatrix} \boldsymbol{r}_p^T & \boldsymbol{r}_d^T \end{bmatrix}^T$, $\boldsymbol{\Psi}_B = \begin{bmatrix} \boldsymbol{\Psi}_p^T & \boldsymbol{\Psi}_d^T \end{bmatrix}^T$ and so on, and $\boldsymbol{r}_0 = \boldsymbol{r}_p$. So
$$
\begin{bmatrix} \boldsymbol{u}_p \\ \boldsymbol{\psi}_p \\ \boldsymbol{u}_d \\ \boldsymbol{\psi}_d \end{bmatrix} = 
\begin{bmatrix}
\boldsymbol{R}_p^T (\boldsymbol{r}_p - \boldsymbol{r}_p) - \boldsymbol{X}_p \\
-\boldsymbol{\Psi}_p \circ \boldsymbol{\Psi}_p \\
\boldsymbol{R}_p^T (\boldsymbol{r}_d - \boldsymbol{r}_p) - \boldsymbol{X}_d \\
-\boldsymbol{\Psi}_d \circ \boldsymbol{\Psi}_d 
\end{bmatrix} =
\begin{bmatrix}
\boldsymbol{0} \\
\boldsymbol{0} \\
\boldsymbol{R}_p^T (\boldsymbol{r}_d - \boldsymbol{r}_p) - \boldsymbol{X}_d \\
-\boldsymbol{\Psi}_d \circ \boldsymbol{\Psi}_d 
\end{bmatrix}
$$

It's hard to continue without assuming that the relative rotations $\boldsymbol{\psi}_B$ are small. From Geradin & Cardona (p. 201), the variation of these deflections are
$$
\delta \boldsymbol{u}_d =
\boldsymbol{R}_p^T (\delta \boldsymbol{r}_d - \delta \boldsymbol{r}_p) +
\left[ \boldsymbol{R}_p^T (\boldsymbol{r}_d - \boldsymbol{r}_p) \right] \times \delta \boldsymbol{\Theta_p}
$$
The first part of this is clear -- a motion of either point changes the relative displacement between then. The $\delta \boldsymbol{\Theta}_p$ appears because a rotation of the proximal node rotates the coordinate system and hence changes $\boldsymbol{u}_d$. Geradin & Cardona give the variation in the local rotation as
$$
\delta \boldsymbol{\psi}_d = \boldsymbol{T}^{-1}(\boldsymbol{\psi}_d) \left( \delta\boldsymbol{\Theta}_d - \exp{(\tilde{\boldsymbol{\psi}}_d)}^T \delta\boldsymbol{\Theta}_0 \right)
$$
The function $\boldsymbol{T}(\beta)$ is the "tangent operator", which gives the angular velocity vector from the derivatives of the rotation parameters; for small relative rotations, its inverse is approximated by $\boldsymbol{I} + \tilde{\boldsymbol{\psi}}/2$ if $\boldsymbol{\psi}$ is the Cartesian rotation vector. I don't quite follow this but, but according to G&C this becomes
$$
\delta \boldsymbol{\psi}_d = 
(2\boldsymbol{I} + \tilde{\boldsymbol{\psi}_d})/2 \delta\boldsymbol{\Theta}_d - 
(2\boldsymbol{I} + \tilde{\boldsymbol{\psi}_d})/2 \delta\boldsymbol{\Theta}_p
$$

The variation in the coordinates $\boldsymbol{y}$ is 
$$\delta \boldsymbol{y} = \boldsymbol{Y} \delta\boldsymbol{q} = 
\begin{bmatrix} \boldsymbol{Y}_B & \boldsymbol{0} \\ \boldsymbol{0} & \boldsymbol{I} \end{bmatrix}
\begin{bmatrix} \delta \boldsymbol{r}_B \\ \delta \boldsymbol{\Theta}_B \\ \boldsymbol{\eta} \end{bmatrix}
$$ 

The first two rows of $\boldsymbol{Y}_B$ is zero, as there is no deflection or rotation at the proximal node. The third and fourth rows are
$$
\boldsymbol{Y}_{d} = \begin{bmatrix} 
-\boldsymbol{R}_p^T & \boldsymbol{0} & \boldsymbol{R}_p^T & \boldsymbol{0} \\
\boldsymbol{0} & -(2\boldsymbol{I} + \tilde{\boldsymbol{\psi}_d})/2 &
\boldsymbol{0} &  (2\boldsymbol{I} + \tilde{\boldsymbol{\psi}_d})/2
\end{bmatrix}
$$


Rewriting this to depend on the generalised coordinates,
$$
\begin{bmatrix} \boldsymbol{u}_p \\ \boldsymbol{\psi}_p \\ \boldsymbol{u}_d \\ \boldsymbol{\psi}_d \end{bmatrix} = 
\begin{bmatrix}
\boldsymbol{0} & \boldsymbol{0} & \boldsymbol{0} & \boldsymbol{0} \\
\boldsymbol{0} & \boldsymbol{0} & \boldsymbol{0} & \boldsymbol{0} \\
-\boldsymbol{R}_p^T & \boldsymbol{0} & \boldsymbol{R}_p^T & \boldsymbol{0} \\
-\boldsymbol{\Psi}_d \circ \boldsymbol{\Psi}_d 
\end{bmatrix}
\begin{bmatrix} \boldsymbol{r}_p \\ \boldsymbol{\Psi}_p \\ \boldsymbol{r}_d \\ \boldsymbol{\Psi}_d \end{bmatrix}
+
\begin{bmatrix} \boldsymbol{0} \\ \boldsymbol{0} \\ -\boldsymbol{X}_d \\  \end{bmatrix}
$$
