To use the Craig-Bampton method, two things are needed:

  1) the static boundary modes, $\boldsymbol{\Xi} = 
    \begin{bmatrix} \boldsymbol{I} \\ -\boldsymbol{K}_{II}^{-1}\boldsymbol{K}_{IB} \end{bmatrix}$
    
  2) a reduced set of $r$ normal modes, $\boldsymbol{\Phi}$
    

# Step 1 - FE model of beam

Test the FE model in mbwind. This model matches the Bladed file "demo_a_uniform_blade".

```python
from mbwind.modes import ModesFromScratch
x = arange(0, 61, 3)
modes = ModesFromScratch(x, 1000, 1, 5e10, 5e10)
matshow(modes.K); colorbar()
matshow(modes.M); colorbar();
```

## Normal modes

First, calculate the normal modes with both ends of the beam clamped.

```python
wn, vn = modes.normal_modes(clamped=True)
```

```python
def plot_mode(i, v, w=None):
    fig, ax = plt.subplots(2, sharex=True, figsize=(4,2))
    xvals = arange(0, 61, 3)
    ax[0].plot(xvals, v[1::6, i], xvals, v[2::6, i], xvals, v[3::6, i])
    ax[0].grid(True)
    ax[0].set_ylabel('Deflection')
    ax[1].plot(xvals, v[4::6, i], xvals, v[5::6, i], xvals, v[3::6, i])
    ax[1].grid(True)
    ax[1].set_ylabel('Rotation')
    if w is not None:
        ax[0].set_title('Mode %d: %.3f Hz' % (i, w[i]/2/pi))
    else:
        ax[0].set_title('Mode %d: ????? Hz' % i)
        
for i in range(6):
    plot_mode(i, vn, wn)
```

In fact, let's just take the first 3 normal modes:

```python
wn, Phi = modes.normal_modes(n_modes=3)
Phi.shape
```

## Attachment modes

Calculate the attachment modes by applying a unit deflection to each boundary DOF in turn:

```python
Xi = modes.attachment_modes()
```

```python
for i in range(Xi.shape[1]):
    plot_mode(i, Xi)
```

# Step 2 - substructuring

Combined, the attachment and normal modes give the Craig-Bampton reduction matrix:

```python
CB_reduction = c_[ Xi, Phi ]
```

From this we get the reduced stiffness and mass matrices:

```python
Kred = dot(CB_reduction.T, dot(modes.K, CB_reduction))
Mred = dot(CB_reduction.T, dot(modes.M, CB_reduction))
matshow(Kred)
matshow(Mred);
```

Check that the internal parts of the reduced matrices recover the natural frequencies:

```python
allclose(sqrt(diag(Kred[12:, 12:]) / diag(Mred[12:, 12:])), wn[:3])
```

Also, we expect that there is no stiffness coupling, but there is inertial coupling, between the internal modes and the boundary DOFs:

```python
amax(abs(Kred[:12, 12:])) / amax(abs(Kred))
```

# Step 4 - Coordinate transformations

The reduced matrices I have derived above are defined in terms of the local coordinates $\boldsymbol{y}$:
$$\boldsymbol{y} = \begin{bmatrix} \boldsymbol{u}_B \\ \boldsymbol{\psi}_B \\ \boldsymbol{\eta} \end{bmatrix}$$
(although not exactly in that order -- the deflections and rotations are interleaved). I'm assuming small rotations within the body frame, so I don't have to be so precise about how the rotations $\boldsymbol{\psi}_B$ are defined -- they are just small rotations about each axis.

The deflection velocity and angular velocity at each boundary node are found by writing the standard expressions for the velocity of a point in global coordinates, then rearranging to give:
$$
\begin{bmatrix} \dot{\boldsymbol{u}}_B \\ \boldsymbol{\Omega}_B \end{bmatrix} = 
\begin{bmatrix} \boldsymbol{R}_0^T (\dot{\boldsymbol{r}}_B - \dot{\boldsymbol{r}}_0) +
  \boldsymbol{R}_0^T \widetilde{(\boldsymbol{r}_B - \boldsymbol{r}_0)}\boldsymbol{\omega}_0 \\
\boldsymbol{R}_0^T (\boldsymbol{\omega}_B - \boldsymbol{\omega}_0) \end{bmatrix}
$$
where the subscipt $0$ refers to the reference frame. Here I'm taking the proximal node as the reference frame. So
$$
\begin{bmatrix} \dot{\boldsymbol{u}}_p \\ \boldsymbol{\Omega}_p \\ \dot{\boldsymbol{u}}_d \\ \boldsymbol{\Omega}_d \end{bmatrix} = 
\begin{bmatrix}
\boldsymbol{0} & \boldsymbol{0} & \boldsymbol{0} & \boldsymbol{0} \\
\boldsymbol{0} & \boldsymbol{0} & \boldsymbol{0} & \boldsymbol{0} \\
-\boldsymbol{R}_p^T & \boldsymbol{R}_p^T \widetilde{(\boldsymbol{r}_d - \boldsymbol{r}_p)} & \boldsymbol{R}_p^T & \boldsymbol{0} \\
\boldsymbol{0} & -\boldsymbol{R}_p^T & \boldsymbol{0} & \boldsymbol{R}_p^T \\
\end{bmatrix}
\begin{bmatrix} \dot{\boldsymbol{r}}_p \\ \boldsymbol{\omega}_p \\ \dot{\boldsymbol{r}}_d \\ \boldsymbol{\omega}_d \end{bmatrix}
$$
and, calling this matrix above $\boldsymbol{Y}$,
$$
\boldsymbol{\dot{y}} = \begin{bmatrix}
  \boldsymbol{Y} & \boldsymbol{0} \\
  \boldsymbol{0} & \boldsymbol{I} \end{bmatrix}
\begin{bmatrix}
  \dot{\boldsymbol{r}}_p \\
  \boldsymbol{\omega}_p \\
  \dot{\boldsymbol{r}}_d \\
  \boldsymbol{\omega}_d \\
  \boldsymbol{\dot{\eta}} 
\end{bmatrix}
$$

(this is checked in unit tests)

Now make the transformation matrix for the case when the proximal node is at 0, the distal node is 60 along the x-axis, and transform the reduced matrices into the global coordinates:

```python
Y = zeros((15, 15))
Y[:12, :12] = modes.transformation_to_global_coords(eye(3), zeros(3), array([60,0,0]))
Y[12:, 12:] = eye(3)
Kredg = dot(Y.T, dot(Kred, Y))
Mredg = Mred #dot(Y.T, dot(Mred, Y))
```

## Modal mass

Bladed v4.4 now reports the "modal mass" and "modal stiffness". Geradin & Cardona define the modal mass as 
$$
\boldsymbol{M}_i = \frac{\boldsymbol{h}_i \boldsymbol{h}_i^T}{\omega_i^4 \mu_i}
$$
with the "modal participation" being
$$
\boldsymbol{h}_i = (\boldsymbol{K}_{BI} - \omega_i^2 \boldsymbol{M}_{BI})\phi_i
$$

```python
modal_participation = dot((modes.K_BI - wn[0]**2 * modes.M_BI), Phi[6:-6, 0])
```

```python
modal_participation / 1e6
```

```python
M1 = dot(c_[modal_participation], c_[modal_participation].T) / (wn[0]**4 * dot(Phi[:,0].T, dot(modes.M, Phi[:,0])))
(M1/1e3).astype(int)
```

This is much higher than the value from Bladed...
